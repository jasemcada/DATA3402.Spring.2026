```python
class counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.count = 0
    
    def increment(self):
        if self.count < self.max_value:
            self.count += 1
        else:
            print("Error: Cannot increment beyond maximum value")
    
    def reset(self):
        self.count = 0
```

```python
import math

class shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y
    
    def area(self):
        pass
    
    def perimeter(self):
        pass
    
    def get_x(self):
        return self._x
    
    def get_y(self):
        return self._y
    
    def points_on_perimeter(self, n=16):
        pass


class rectangle(shape):
    def __init__(self, length, width, x, y):
        super(rectangle, self).__init__(x, y)
        self.__length = length
        self.__width = width
    
    def area(self):
        return self.__length * self.__width
    
    def perimeter(self):
        return 2 * self.__length + 2 * self.__width
    
    def get_length(self):
        return self.__length
    
    def get_width(self):
        return self.__width
    
    def points_on_perimeter(self, n=16):
        points = []
        total_perimeter = self.perimeter()
        step = total_perimeter / n
        
        for i in range(n):
            distance = i * step
            
            if distance < self.__length:
                px = self._x + distance
                py = self._y
            elif distance < self.__length + self.__width:
                px = self._x + self.__length
                py = self._y + (distance - self.__length)
            elif distance < 2 * self.__length + self.__width:
                px = self._x + self.__length - (distance - self.__length - self.__width)
                py = self._y + self.__width
            else:
                px = self._x
                py = self._y + self.__width - (distance - 2 * self.__length - self.__width)
            
            points.append((px, py))
        
        return points


class circle(shape):
    def __init__(self, radius, x, y):
        super(circle, self).__init__(x, y)
        self.__radius = radius
    
    def area(self):
        return math.pi * self.__radius * self.__radius
    
    def perimeter(self):
        return 2 * math.pi * self.__radius
    
    def get_radius(self):
        return self.__radius
    
    def points_on_perimeter(self, n=16):
        points = []
        
        for i in range(n):
            angle = 2 * math.pi * i / n
            px = self._x + self.__radius * math.cos(angle)
            py = self._y + self.__radius * math.sin(angle)
            points.append((px, py))
        
        return points


class triangle(shape):
    def __init__(self, side_a, side_b, side_c, x, y):
        super(triangle, self).__init__(x, y)
        self.__side_a = side_a
        self.__side_b = side_b
        self.__side_c = side_c
    
    def area(self):
        s = (self.__side_a + self.__side_b + self.__side_c) / 2
        return math.sqrt(s * (s - self.__side_a) * (s - self.__side_b) * (s - self.__side_c))
    
    def perimeter(self):
        return self.__side_a + self.__side_b + self.__side_c
    
    def get_side_a(self):
        return self.__side_a
    
    def get_side_b(self):
        return self.__side_b
    
    def get_side_c(self):
        return self.__side_c
    
    def points_on_perimeter(self, n=16):
        points = []
        total_perimeter = self.perimeter()
        step = total_perimeter / n
        
        x1, y1 = self._x, self._y
        x2, y2 = self._x + self.__side_a, self._y
        x3, y3 = self._x + self.__side_a / 2, self._y + math.sqrt(self.__side_a ** 2 - (self.__side_a / 2) ** 2)
        
        for i in range(n):
            distance = i * step
            
            if distance < self.__side_a:
                t = distance / self.__side_a
                px = x1 + t * (x2 - x1)
                py = y1 + t * (y2 - y1)
            elif distance < self.__side_a + self.__side_b:
                t = (distance - self.__side_a) / self.__side_b
                px = x2 + t * (x3 - x2)
                py = y2 + t * (y3 - y2)
            else:
                t = (distance - self.__side_a - self.__side_b) / self.__side_c
                px = x3 + t * (x1 - x3)
                py = y3 + t * (y1 - y3)
            
            points.append((px, py))
        
        return points
```

```python
import math

class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]

    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]

    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class shape:
    def __init__(self, x, y):
        self._x = x
        self._y = y

    def area(self):
        raise NotImplementedError

    def perimeter(self):
        raise NotImplementedError

    def points_on_perimeter(self, n=32):
        raise NotImplementedError

    def is_inside(self, x, y):
        raise NotImplementedError

    def paint(self, canvas, char='*'):
        points = self.points_on_perimeter(32)
        for px, py in points:
            row = int(py)
            col = int(px)
            canvas.set_pixel(row, col, char=char)

class rectangle(shape):
    def __init__(self, length, width, x, y):
        super(rectangle, self).__init__(x, y)
        self.__length = length
        self.__width = width

    def area(self):
        return self.__length * self.__width

    def perimeter(self):
        return 2 * self.__length + 2 * self.__width

    def points_on_perimeter(self, n=32):
        points = []
        total = self.perimeter()
        step = total / n

        for i in range(n):
            d = i * step

            if d < self.__length:
                px = self._x + d
                py = self._y
            elif d < self.__length + self.__width:
                px = self._x + self.__length
                py = self._y + (d - self.__length)
            elif d < 2 * self.__length + self.__width:
                px = self._x + self.__length - (d - self.__length - self.__width)
                py = self._y + self.__width
            else:
                px = self._x
                py = self._y + self.__width - (d - 2 * self.__length - self.__width)

            points.append((px, py))

        return points

    def is_inside(self, x, y):
        return (self._x <= x <= self._x + self.__length and
                self._y <= y <= self._y + self.__width)

    def __repr__(self):
        return f"rectangle({self.__length}, {self.__width}, {self._x}, {self._y})"

class circle(shape):
    def __init__(self, radius, x, y):
        super(circle, self).__init__(x, y)
        self.__radius = radius

    def area(self):
        return math.pi * self.__radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.__radius

    def points_on_perimeter(self, n=32):
        points = []

        for i in range(n):
            angle = 2 * math.pi * i / n
            px = self._x + self.__radius * math.cos(angle)
            py = self._y + self.__radius * math.sin(angle)
            points.append((px, py))

        return points

    def is_inside(self, x, y):
        dx = x - self._x
        dy = y - self._y
        return dx * dx + dy * dy <= self.__radius ** 2

    def __repr__(self):
        return f"circle({self.__radius}, {self._x}, {self._y})"

class triangle(shape):
    def __init__(self, side, x, y):
        super(triangle, self).__init__(x, y)
        self.__side = side

    def area(self):
        return (math.sqrt(3) / 4) * self.__side ** 2

    def perimeter(self):
        return 3 * self.__side

    def points_on_perimeter(self, n=32):
        points = []
        total = self.perimeter()
        step = total / n

        x1, y1 = self._x, self._y
        x2, y2 = self._x + self.__side, self._y
        x3 = self._x + self.__side / 2
        y3 = self._y + math.sqrt(3) / 2 * self.__side

        for i in range(n):
            d = i * step

            if d < self.__side:
                t = d / self.__side
                px = x1 + t * (x2 - x1)
                py = y1 + t * (y2 - y1)
            elif d < 2 * self.__side:
                t = (d - self.__side) / self.__side
                px = x2 + t * (x3 - x2)
                py = y2 + t * (y3 - y2)
            else:
                t = (d - 2 * self.__side) / self.__side
                px = x3 + t * (x1 - x3)
                py = y3 + t * (y1 - y3)

            points.append((px, py))

        return points

    def is_inside(self, x, y):
        x1, y1 = self._x, self._y
        x2, y2 = self._x + self.__side, self._y
        x3 = self._x + self.__side / 2
        y3 = self._y + math.sqrt(3) / 2 * self.__side

        def area(a, b, c, d, e, f):
            return abs((a*(d-f) + c*(f-b) + e*(b-d)) / 2)

        A = self.area()
        A1 = area(x, y, x2, y2, x3, y3)
        A2 = area(x1, y1, x, y, x3, y3)
        A3 = area(x1, y1, x2, y2, x, y)

        return abs(A - (A1 + A2 + A3)) < 1e-5

    def __repr__(self):
        return f"triangle({self.__side}, {self._x}, {self._y})"

class CompoundShape(shape):
    def __init__(self):
        self._shapes = []

    def add(self, s):
        self._shapes.append(s)

    def area(self):
        total = 0
        for s in self._shapes:
            total += s.area()
        return total

    def perimeter(self):
        total = 0
        for s in self._shapes:
            total += s.perimeter()
        return total

    def points_on_perimeter(self, n=32):
        points = []
        for s in self._shapes:
            pts = s.points_on_perimeter(n)
            for p in pts:
                points.append(p)
        return points

    def is_inside(self, x, y):
        for s in self._shapes:
            if s.is_inside(x, y):
                return True
        return False

    def paint(self, canvas, char='*'):
        for s in self._shapes:
            s.paint(canvas, char)
```

```python
canvas = Canvas(60, 25)

r = rectangle(15, 6, 5, 5)
c = circle(5, 35, 10)
t = triangle(10, 20, 15)

comp = CompoundShape()
comp.add(r)
comp.add(c)
comp.add(t)

comp.paint(canvas, char='*')

canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self._shapes = []

    def add(self, shape):
        self._shapes.append(shape)

    def remove(self, shape):
        if shape in self._shapes:
            self._shapes.remove(shape)

    def clear(self):
        self.canvas.clear_canvas()

    def paint(self):
        self.canvas.clear_canvas()
        for s in self._shapes:
            s.paint(self.canvas)
        self.canvas.display()
```

```python
drawing = RasterDrawing(30, 15)

r = rectangle(10, 4, 5, 5)
drawing.add(r)

print("First paint:")
drawing.paint()

drawing.remove(r)
c = circle(4, 15, 7)
drawing.add(c)

print("\nSecond paint:")
drawing.paint()
```

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self._shapes = []

    def add(self, shape):
        self._shapes.append(shape)

    def remove(self, shape):
        if shape in self._shapes:
            self._shapes.remove(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for s in self._shapes:
            s.paint(self.canvas)
        self.canvas.display()

    def __repr__(self):
        rep = f"drawing = RasterDrawing({self.canvas.width}, {self.canvas.height})\n"
        for s in self._shapes:
            rep += f"drawing.add({repr(s)})\n"
        return rep

    def save(self, filename):
        with open(filename, "w") as f:
            f.write(self.__repr__())
```

```python
def load_drawing(filename):
    with open(filename, "r") as f:
        code = f.read()

    local_vars = {}
    exec(code, globals(), local_vars)

    return local_vars["drawing"]
```

```python
drawing = RasterDrawing(40, 20)

r = rectangle(10, 4, 5, 5)
c = circle(5, 25, 10)
t = triangle(8, 15, 3)

drawing.add(r)
drawing.add(c)
drawing.add(t)

print("Original drawing:")
drawing.paint()

drawing.save("test_drawing.txt")

loaded = load_drawing("test_drawing.txt")

print("\nLoaded drawing:")
loaded.paint()
```
